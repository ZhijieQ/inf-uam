```python
# Cortes
def cortes(zonas_picos):
    cuts = []
    i1 = 0
    while True:
        if i1 >= len(zonas_picos)-1:
            break
        if zonas_picos[i1] == False and zonas_picos[i1+1] == True:
            for i2 in range(i1+1, len(zonas_picos)-1):
                if zonas_picos[i2] == True and zonas_picos[i2+1] == False:
                    cuts.append((i1,i2))
                    i1 = i2
                    break
                    
        i1 = i1 + 1
    print cuts
    cuts = map(lambda (i0,i1): (i0[0]+5,i1[1]-5), zip(cuts[::2],cuts[1::2]))
    print cuts

    return cuts

cutsH = cortes(horizontal>thrsH)
cutsV = cortes(vertical>thrsV)
```

```python
def extrae_caracs(numerito):
    filas, columnas = numerito.imagen.shape
    paso_fil = filas / 20.
    paso_col = columnas / 12.
    caracteristicas = []
    ff = np.arange(0, filas,    paso_fil)
    ff = np.append(ff, filas)
    cc = np.arange(0, columnas, paso_col)
    cc = np.append(cc, columnas)
    for f in range(0,len(ff)-2):
        for c in range(0,len(cc)-2):
            caracteristicas.append(numerito.imagen[ff[f]:ff[f+2],cc[c]:cc[c+2]].sum())

    return caracteristicas

c = extrae_caracs(num)
print c
```
